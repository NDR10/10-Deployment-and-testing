```python
import sklearn
import pandas as pd
```

```python
dataset = pd.read_csv ("50_Startups.csv")
dataset
```

```python
dataset.columns
```

```python
dataset.shape
```

```python
dataset.info()
dataset.describe().T
```

```python
dataset = pd.get_dummies(dataset, dtype = int, drop_first = True)
dataset
```

```python
Input = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
Input
```

```python
Output = dataset[["Profit"]]
Output
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, Y_train, Y_test = tts (Input, Output, test_size = 0.3, random_state = 0)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeRegressor as DTR
Model = DTR(criterion = "absolute_error", splitter= "random")
Model = Model.fit(X_train, Y_train)
```

```python
X_train
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(Model)
plt.show()
```

```python
Y_pred = Model.predict(X_test)
```

```python
Y_pred = Model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(Y_test, Y_pred)
r_score
```
